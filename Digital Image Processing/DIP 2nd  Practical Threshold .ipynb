```python
def threshold(image):
    for row in range(image.shape[0]):
        for col in range(image.shape[1]):
            if (image[row,col,0])>200:
                red=1
            else:
                red=0
            if (image[row,col,1])>200:
                green=1
            else:
                green=0
            if (image[row,col,2])>200:
                blue=1
            else:
                blue=0
            i[row,col]=[red,green,blue]
    return i
```
